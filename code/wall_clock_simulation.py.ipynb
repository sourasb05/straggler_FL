```python
import random
import numpy as np
import os
import pickle
```

```python
 path_project = os.path.abspath('..')
```

```python
idxs_users_wc = []
```

```python
idxs_users_wc = random.sample(range(990, 2000), 90)

idxs_users_wc.extend(random.sample(range(2000,10000), 10))
print("idxs_users_wc",idxs_users_wc)
```

```python
random.shuffle(idxs_users_wc)
print(idxs_users_wc)
```

```python
max_time_random = np.array([])
print(max_time_random)
```

```python
for i in range(100): 
    sample = random.sample(idxs_users_wc, 10)
    #print("sample",sample)
    max_delay = np.amax(sample)
    print("max_delay",max_delay)
    max_time_random = np.append(max_time_random,max_delay)
print("max_time_random",max_time_random)
```

```python
max_time_fedmoods = np.array([])
```

```python
idxs_users_wc.sort()
#idxs_users_wc
```

```python
print(max_time_fedmoods)
n = 5
k = 1
for i in range(100):
    if n <= len(idxs_users_wc):
        sample = idxs_users_wc[0:n]
        #print(sample)
        n=n+k
    else:
        sample = idxs_users_wc[0:n]
        print("here I am")
        #print(sample)
    
    max_delay = np.amax(sample)
    print(max_delay)
    max_time_fedmoods = np.append(max_time_fedmoods,max_delay)
    print("max_time_fedmoods",max_time_fedmoods)
```

```python
file_name = path_project + '/save/Fed-MOODS-WC_-FMNIST-straggler[{}]_adaptive[{}].pkl'. \
            format(10, 1)

with open(file_name, 'wb') as f:
    pickle.dump([max_time_random, max_time_fedmoods],f)
```
